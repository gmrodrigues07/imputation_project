```python
import torch
from torch.utils.data import Dataset, DataLoader
class NumpyDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data.astype(np.float32)
        self.labels = labels.reshape(-1,1)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = torch.from_numpy(self.data[idx])
        label = torch.from_numpy(self.labels[idx])
        return sample, label[0]


# Dataset ready to work on PyTorch    
dataset_train = NumpyDataset(X_train, y_train)
dataset_test = NumpyDataset(X_test, y_test)
```

Using Pytorch, complete the cell below to implement a model that follows the architecture previously described.

```python
import torch
from torch.utils.data import Dataset, DataLoader

class DenseNetwork(torch.nn.Module):
    def __init__(self):
        super().__init__()
        # Layer 1: Input (size 10 based on data) -> Dense 64
        self.layer1 = torch.nn.Linear(10, 64)
        # Layer 2: Dense 64 -> Dense 32
        self.layer2 = torch.nn.Linear(64, 32)
        # Layer 3: Dense 32 -> Dense 16
        self.layer3 = torch.nn.Linear(32, 16)
        # Output Layer: Dense 16 -> Output 3
        self.output_layer = torch.nn.Linear(16, 4)
        self.relu = torch.nn.ReLU()

    def forward(self, x):
        # Apply layer 1 + ReLU
        x = self.relu(self.layer1(x))
        # Apply layer 2 + ReLU
        x = self.relu(self.layer2(x))
        # Apply layer 3 + ReLU
        x = self.relu(self.layer3(x))
        # Apply output layer (returning raw logits, no activation)
        x = self.output_layer(x)
        return x

network = DenseNetwork()
network
```

#### 2 - c) 

Write the training loop to fit the network you previously defined. Your function must receive as inputs the model, the training loader, the test loader, the loss function (criterion), the optimizer, and the number of epochs. You should also return the training and testing accuracy and loss for each epoch. Do this by completing the following cell.

```python
def train_model(model, train_loader, test_loader, criterion, optimizer, num_epochs=10):
    train_acc = []
    train_loss = []

    test_acc = []
    test_loss = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for x, y in train_loader:
            optimizer.zero_grad()
            outputs = model(x)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * x.size(0)
            _, preds = torch.max(outputs, 1)
            correct += (preds == y).sum().item()
            total += y.size(0)

        train_loss.append(running_loss / total)
        train_acc.append(correct / total)

        model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for x, y in test_loader:
                outputs = model(x)
                loss = criterion(outputs, y)

                running_loss += loss.item() * x.size(0)
                _, preds = torch.max(outputs, 1)
                correct += (preds == y).sum().item()
                total += y.size(0)

        test_loss.append(running_loss / total)
        test_acc.append(correct / total)

    return train_acc, train_loss, test_acc, test_loss
```

#### 2 - d)

Using a batch size of 8, train your model with the Stochastic Gradient Descent optimizer and a learning rate of 1e-2, using the cross-entropy loss function for 10 epochs. Next, plot the evolution of training and testing accuracy and loss.

```python
import matplotlib.pyplot as plt


train_loader = DataLoader(dataset_train, batch_size=8, shuffle=True)
test_loader = DataLoader(dataset_test, batch_size=8, shuffle=False)

num_epochs = 10
opt = torch.optim.SGD(network.parameters(), lr=1e-2)
loss_fn = torch.nn.CrossEntropyLoss()

train_acc, train_loss, test_acc, test_loss = train_model(model=network, train_loader=train_loader, test_loader=test_loader, criterion=loss_fn, optimizer=opt, num_epochs=num_epochs)

epochs = range(1, num_epochs + 1)

plt.figure()
plt.plot(epochs, train_acc, label="Train Accuracy")
plt.plot(epochs, test_acc, label="Test Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.figure()
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, test_loss, label="Test Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
